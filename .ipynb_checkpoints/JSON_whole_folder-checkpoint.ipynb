```python
# make a list of names with .json in in path
dic_all = {}
for file in os.listdir(PATH_in):
    if file.endswith(".xml"):
        #print(file)
        dic = XMLtoJSON.XMLtoJSON(Path_xml = os.path.join(PATH_in, file), buffer = 30)
        dic_all.update(dic)
        """
        try:
            dic = XMLtoJSON.XMLtoJSON(input = os.path.join(PATH_in, file), buffer = 30)
            dic_all.update(dic)
        except:
            print('This failed:',file)
            pass
        """
len(dic_all)
```
