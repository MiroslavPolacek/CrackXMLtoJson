```python
import os
import shutil
import xml.etree.ElementTree as ET

#xml_example = 'XMLexample/Alex_20191216.xml'
```

## Move all images in one folder to simplify upload

```python
image_path = '/Volumes/MiroAllLife/Annotations/Tasks'
try:
    os.mkdir('/Volumes/MiroAllLife/Annotations/Tasks_unites')
except:
    pass
# go find all tiffs in subdirectories of this folder and move them to Tasks_united
for dirpath, dirnames, filenames in os.walk(image_path):
    for filename in [f for f in filenames if f.endswith(".tif")]:
        shutil.copy(os.path.join(dirpath, filename), '/Volumes/MiroAllLife/Annotations/Tasks_unites')
```

## Merge all annotations in one xml

```python
xml_path = '/Volumes/MiroAllLife/Annotations/AnnotationsXMLCorrected100320'
#xml_path = 'XMLexample'
xml_files = os.listdir(xml_path)
print(xml_files)
the_first = ET.parse(os.path.join(xml_path, xml_files[0])).getroot()
print(len(the_first))

for file in xml_files[1:]:
    print(file)
    root = ET.parse(os.path.join(xml_path, file)).getroot()
    
    for child in root:
        if child.tag == 'image':
            print(child.attrib)
            the_first.append(child)
            
print(len(the_first))
#the_first.write('united_xml.xml')
mydata = ET.tostring(the_first)
myfile = open("united_xml.xml", "wb")
myfile.write(mydata)
```
